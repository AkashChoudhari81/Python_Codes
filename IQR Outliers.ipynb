```python
Q1 = np.percentile(df.Marks,25,interpolation = 'midpoint')
Q3 = np.percentile(df.Marks,75,interpolation = 'midpoint')
```

```python
IQR = Q3-Q1
print(IQR)
```

```python
upper_bound = Q3 + (1.5*IQR)
lower_bound = Q1 - (1.5*IQR)
print(upper_bound)
print(lower_bound )
```

```python
upper_limit =df.Marks>=upper_bound
lower_limit =df.Marks<=lower_bound
print(upper_limit)
print(lower_limit)
```

```python
outliers = []
for x in range(len(dataset['Marks'])):
    if (upper_limit[x] == True) or (lower_limit[x] == True):
        outliers.append(dataset['Marks'][x])

print(outliers)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import mean_squared_error,f1_score,accuracy_score,precision_score
```

```python
df.columns
```

```python
x = df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
y = df[['MEDV']]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25, random_state = 45)
```

```python
y_pred = model.predict(x_test)
```

```python
print("Predictions are :",y_pred)
```

```python
f1_score = model.score(x_train,y_train)
```

```python
print("F1_score is :",f1_score)
```

```python
from sklearn.metrics import classification_report
```

```python
Y_pred = np.argmax(model.predict(x_test), axis = 0)
```

```python
print(Y_pred)
```

```python
print(confusion_matrix(y_test,y_pred))
```
